```python
!pip install matplotlib
import time
import requests
import matplotlib.pyplot as plt

my_list = []

for i in range(102):
    print(i)
    res = requests.get('http://api.open-notify.org/iss-now.json')
    json_data = res.json()
    my_list.append(json_data)
    time.sleep(10)
```

```python
x=[]
for i in range(0,102):
    latitude = my_list[i].get('iss_position',{}).get('latitude',[])
    x.append(latitude)
    
print(x)
    
y=[]
for j in range(0,102):
    longitude = my_list[j].get('iss_position',{}).get('longitude',[])
    y.append(longitude)

print(y)

plt.plot(x, y)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot of the ISS space station (17 mins)')
plt.ylim(min(y), max(y))
plt.show()
```
